```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
sm=pd.read_csv('marks_1.csv')
sm
```

```python
sm['Total']=sm[['English','Hindi','Science','SST']].sum(axis=1)
sm
```

```python
sm['Percentage']=(sm['Total']/5)
sm
```

```python
A=100-80
B=80-60
C=60-50
D=50-40
E=40-30
F=0-30
```

```python
def clac_grade(p):
    if p<=100 and p>=80:
        return 'A'
    elif p<80 and p>=60:
        return 'B'
    elif p<60 and p>=50:
        return 'C'
    elif p<50 and p>=40:
        return 'D'
    elif p<40 and p>=30:
        return 'E'
    elif p<30 and p>=0:
        return 'F'
    else:
       return 'X'

sm['Grade']=sm['Percentage'].apply(clac_grade)
sm
```
